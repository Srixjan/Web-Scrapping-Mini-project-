# 2. Read .csv from Multiple URLs with Pandas

```python
 root = 'https://www.football-data.co.uk/mmz4281/'
```

```python
leagues = ['E0','E2','E3']
frames = []

for league in leagues:
    df = pd.read_csv( root + '2425' + '/' + league + '.csv')
    frames.append(df)
```

```python
leagues = ['E0','E2','E3']
frames = []

for league in leagues:
    for season in range(15, 24):
        df = pd.read_csv( root + str(season)+str(season+1) + '/' + league + '.csv')
        df.insert(1, 'season', season)
        frames.append(df)
```

# Organize All the Data in a Dictionary

```python
dict_countries = {'Spanish':'SP1','Spanish Segunda Division':'SP2','English Premier League':'E0','German':'D1','English League':'E2'}
```

```python
dict_countries
```

```python
dict_countries['English Premier League']
```

```python
for leagues in dict_countries:
    print(dict_countries[leagues])
```

```python
dict_historical_data = {}
for league in dict_countries:
    frames = []
    for season in range(15, 24):
        df = pd.read_csv( root + str(season)+str(season+1) + '/' + dict_countries[leagues] + '.csv')
        df.insert(1, 'season', season)
        frames.append(df)
    df_concat =  pd.concat(frames)
    dict_historical_data[league] = df_concat
```
